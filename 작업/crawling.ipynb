```python
from urllib import parse
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import csv

browser = webdriver.Chrome(r'C:/Program Files/Google/Chrome/Application/Chrome.exe') 
```

```python
website = '블라인드'  # 검색하는 곳의 커뮤니티 명
company = '카카오뱅크'  # 검색하고자 하는 기업명 입력
company_url = parse.quote(company)
p = 0

f = open(f'C:/Users/gmkim/Documents/GitHub/JWAIP/경민_작업기록/{website}_{company}.csv', 'w', encoding='euc-kr', newline='')
csvWriter = csv.writer(f)
csvWriter.writerow(['reviews'])  # key 명 지정
```

```python
# case 1
# 1 ~ 마지막 페이지 크롤링
while True:
    p += 1
    url = f'https://www.teamblind.com/kr/company/{company_url}/reviews?page={p}'
    browser.get(f'{url}')
    time.sleep(2)

    elems = browser.find_elements_by_class_name('rvtit')
    for j in elems:
        print(j.text)
        csvWriter.writerow([j.text])
        
    browser.find_element_by_css_selector('body').send_keys(Keys.END)
    time.sleep(2)

    if not elems:
        break
```
